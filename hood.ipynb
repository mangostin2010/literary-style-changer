```python
import re
import time

def limit_repeated_string(text, target_string, max_repeats):
    # 정규 표현식 패턴 생성
    pattern = f"({re.escape(target_string)}){{1,{max_repeats}}}"
    # 패턴과 일치하는 부분만 허용
    return re.sub(f"{re.escape(target_string)}+", lambda m: m.group(0)[:max_repeats * len(target_string)], text)

def gen(x):
    # 모델을 GPU로 이동 (가능한 경우)
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    # 입력 텐서를 GPU로 이동
    inputs = tokenizer(
        f"### 원본: {x}\n\n### 결과:",
        return_tensors='pt',
        return_token_type_ids=False
    )
    inputs = {key: value.to(device) for key, value in inputs.items()}

    # Start time measurement
    start_time = time.time()

    # 텍스트 생성
    gened = model.generate(
        **inputs,
        max_new_tokens=20,
        do_sample=True,
        pad_token_id=tokenizer.eos_token_id  # 경고 제거
    )

    # End time measurement
    end_time = time.time()

    output_text = tokenizer.decode(gened[0])
    result = output_text.split("### 결과:")[1].strip()
    print(limit_repeated_string(result, ';', 3))

    # Print the time taken for response
    print(f"Response time: {end_time - start_time:.4f} seconds")

while True:
    gen(input('Enter: '))
```
