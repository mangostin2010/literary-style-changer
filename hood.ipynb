```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
```

```python
from peft import PeftModel, PeftConfig
```

```python
peft_model_id = "mangostin2010/hood-200step"
config = PeftConfig.from_pretrained(peft_model_id)
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)
model = AutoModelForCausalLM.from_pretrained(config.base_model_name_or_path, quantization_config=bnb_config, device_map={"":0})
model = PeftModel.from_pretrained(model, peft_model_id)
tokenizer = AutoTokenizer.from_pretrained(config.base_model_name_or_path)

model.eval()
```

```python
def gen(x):
    q = f"### 질문: {x}\n\n### 답변:"
    # print(q)
    gened = model.generate(
        **tokenizer(
            q,
            return_tensors='pt',
            return_token_type_ids=False
        ).to('cuda'),
        max_new_tokens=50,
        early_stopping=True,
        do_sample=True,
        eos_token_id=2,
    )
    print(tokenizer.decode(gened[0]))
```

```python
import re

def limit_repeated_string(text, target_string, max_repeats):
    # 정규 표현식 패턴 생성
    pattern = f"({re.escape(target_string)}){{1,{max_repeats}}}"
    # 패턴과 일치하는 부분만 허용
    return re.sub(f"{re.escape(target_string)}+", lambda m: m.group(0)[:max_repeats * len(target_string)], text)

def gen(x):
    # 모델을 GPU로 이동 (가능한 경우)
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    # 입력 텐서를 GPU로 이동
    inputs = tokenizer(
        f"### 원본: {x}\n\n### 결과:",
        return_tensors='pt',
        return_token_type_ids=False
    )
    inputs = {key: value.to(device) for key, value in inputs.items()}

    # 텍스트 생성
    gened = model.generate(
        **inputs,
        max_new_tokens=20,
        do_sample=True,
        pad_token_id=tokenizer.eos_token_id  # 경고 제거
    )
    #print(tokenizer.decode(gened[0]))
    output_text = tokenizer.decode(gened[0])
    result = output_text.split("### 결과:")[1].strip()
    print(limit_repeated_string(result, ';', 3))

while True:
  gen(input('Enter: '))
```
